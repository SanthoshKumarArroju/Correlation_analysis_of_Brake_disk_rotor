# Correlation Analysis of Disk Brake Rotor

```matlab
%% ========================================================== %%
% Modal Analysis on Brake Rotor free-free Boundary conditions  %
% ============================================================ %
% by: Santhosh Kumar Arroju Graduate Student at Wright State University
% used vibration tool box mdofcf for curve fitting and extract mode shapes


load point_1.mat
H1 = Hf_chan_2; 
figure()
% plotting Frequency Response Function 
plot(Freq_domain,20*log10(H1))
xlabel('Freq, Hz')
ylabel('FRF, dB')
```

```matlab
% plotting phase and amplitude
figure()
plot(Freq_domain,rad2deg(angle(H1)))
xlabel('Freq, Hz')
ylabel('Phase, degrees')
figure()
plot(Freq_domain, 20*log10(abs(H1)))
xlabel('Freq, Hz')
ylabel('Amplitude, dB')
```

```matlab
% Energy of Hammer
figure()
plot(Freq_domain, 20*log10(PSD_chan_1))
xlabel('Freq, Hz')
ylabel('PSD, dB')
title('Power Spectral Density of Hammer')
```

```matlab
% plotting Nyquist plot
figure()
plot(real(H1),imag(H1))
xlabel('Real, FRF')
ylabel('Imag, FRF')
```
